# Semantic Segmentation

Build a network from scratch that is able to do semantic segmentaion using the _Semantic Boundaries Dataset_ [Link](https://docs.pytorch.org/vision/0.24/generated/torchvision.datasets.SBDataset.html#torchvision.datasets.SBDataset)

```python
import torch
import torchvision
from torchvision.transforms import v2 as transforms
from IPython.display import display, Image
from pathlib import Path
```

```python
# Data exploration
# SBDataset can only download once, so ensure correct flag is set
sb_train_path = Path("data/sbdata-train")
sb_transforms = transforms.Compose(transforms=[transforms.ToTensor()])
sb_train_ds = torchvision.datasets.SBDataset(root=sb_train_path, image_set="train", download=not sb_train_path.exists(), transforms=sb_transforms)
sb_train_dl = torch.utils.data.DataLoader(dataset=sb_train_ds, batch_size=8, shuffle=True)
for idx, (i1, i2) in enumerate(sb_train_ds):
    print(i1.shape, i2.shape)
    display(transforms.ToPILImage()(i1))
    break
```
